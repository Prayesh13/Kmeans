```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Parameters
X,y = make_blobs(n_samples=300, n_features=2, centers=4, cluster_std=1.5, random_state=42)

# Visulize the Data : 
plt.scatter(X[:,0],X[:,1], s=50)
plt.title("Kmeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_init=10, n_clusters = i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.grid(which='both')
```

```python
kmeans = KMeans(n_init=10, n_clusters=4)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

```python
kmeans.cluster_centers_[:,0]
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1], s=60, c='red', label='cluster1')
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1], s=60, c='blue', label='cluster2')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1], s=60, c='green', label='cluster3')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1], s=60, c='yellow', label='cluster4')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=100, c='black', label='centroid')

plt.title("Kmeans Clustring")
plt.xlabel("F1")
plt.ylabel("F2")
plt.legend()
plt.show()
```
